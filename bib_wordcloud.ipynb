```python
import bibtexparser
from wordcloud import WordCloud
import matplotlib
from matplotlib import pyplot as plt
matplotlib.rcParams['savefig.dpi'] = 400
matplotlib.rcParams["figure.dpi"] = 300
```

```python

with open('acm-papers.bib') as bibtex_file:
    bib_database = bibtexparser.load(bibtex_file)
```

```python
titles = ' '.join([e['title'] for e in bib_database.entries])
abstracts = ' '.join([e['abstract'] for e in bib_database.entries])
```

```python
def plot_wordcloud(text):
    wordcloud = WordCloud(max_font_size=50).generate(text)
    plt.figure(figsize=(15, 20))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
```

```python
plot_wordcloud(titles)
plot_wordcloud(abstracts)
plot_wordcloud(titles + abstracts)
plt.savefig('bib_wordcloud.png', )
```
